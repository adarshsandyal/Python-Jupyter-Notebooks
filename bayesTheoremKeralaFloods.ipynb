```python
impactful_columns = ['YEAR', 'JUN', 'JUL', 'OCT', ' ANNUAL RAINFALL', 'FLOODS']

data = df[impactful_columns]
```

```python
data.head()
```

```python
threshold_jun = data['JUN'].median().astype(int)
threshold_jul = data['JUL'].median().astype(int)
threshold_oct = data['OCT'].median().astype(int)
threshold_ar = data[' ANNUAL RAINFALL'].median().astype(int)

threshold_jun, threshold_jul, threshold_oct, threshold_ar
```

```python
thresholds = {
    'JUN': threshold_jun,
    'JUL': threshold_jul,
    'OCT': threshold_oct,
    ' ANNUAL RAINFALL': threshold_ar
}

for col, threshold in thresholds.items():
    data[col] = (data[col] > threshold).astype(int)

data.head()
```

```python
#Q: What are the chances of Floods when there is a heavy rainfall in June?

#P(Flood == yes | june == 1)
```

```python
pd.crosstab(index = data['JUN'],
            columns = data['FLOODS'],
            margins=True,
            margins_name='Total')
```

```python
# P(flood == yes | june == 1) --> 44/60
```
